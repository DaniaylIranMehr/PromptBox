##### Import necessary libraries

```python
import sys
import gradio as gr
from openai import OpenAI
```

##### create and configure a client object

```python
client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key="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"
)
```

##### call the OpenRouter model and return the text

```python
def model(prompt):
    response = client.chat.completions.create(
    model="openai/gpt-oss-20b:free",
    messages=[
            {
                "role": "user",
                "content": prompt
            }
    ],
    extra_body={"reasoning": {"enabled": True}}
    )

    response = response.choices[0].message.content

    return response
```

##### Create and launch the Gradio interface for user interaction

```python
iface = gr.Interface(
    fn = model,
    inputs = gr.Textbox(label="Enter Text: "),
    outputs = gr.TextArea(label="Generated Answer: "),
    title= "AI-powered model",
    description="Enter your prompt."
)

iface.launch()
```
